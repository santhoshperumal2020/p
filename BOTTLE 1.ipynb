```python
f=c[['Cst_Cnt', 'Btl_Cnt', 'Sta_ID', 'Depth_ID', 'Depthm', 'RecInd',
       'P_qual', 'O_qual', 'O2Satq', 'Chlqua', 'Phaqua', 'PO4q', 'SiO3qu',
       'NO2q', 'NO3q', 'NH3q', 'C14A1q', 'C14A2q', 'DarkAq', 'MeanAq',
       'R_Depth', 'R_PRES']]
```

```python
sns.heatmap(f.corr())
```

```python
x=f[['Cst_Cnt', 'Btl_Cnt','RecInd',
       'P_qual', 'O_qual', 'O2Satq', 'Chlqua', 'Phaqua', 'PO4q', 'SiO3qu',
       'NO2q', 'NO3q', 'NH3q', 'C14A1q', 'C14A2q', 'DarkAq', 'MeanAq',
       'R_Depth', 'R_PRES']]
y=f['Depthm']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
print(lr.intercept_)
```

```python
r=pd.DataFrame(lr.coef_,x.columns,columns=['Co-efficient'])
r
```

```python
u=lr.predict(x_test)
plt.scatter(y_test,u)
```

```python
print(lr.score(x_test,y_test))
```
